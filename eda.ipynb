### Both dataframes have title in common but as matching entries is less than total length some null values will be produced

```python
movies = movies.merge(credits,on='title')
print(movies.shape)
```

```python
null_count=movies.isnull().any(axis=1).sum()
print(null_count)
```

```python
movies.dropna(inplace=True)
```

```python
#Filtering features with significant data
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
# We can use name key for all features but for directors we have to filter the job of Director
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# Only keeping top 5 actors in the movie
movies['cast'] = movies['cast'].apply(convert)
movies['cast'] = movies['cast'].apply(lambda x:x[0:5])
movies.head()
```

```python
movies['cast'].iloc[5]
```

```python
# removing white spaces
def clean(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(clean)
movies['crew'] = movies['crew'].apply(clean)
movies['genres'] = movies['genres'].apply(clean)
movies['keywords'] = movies['keywords'].apply(clean)
```

```python
# movies['overview'] = movies['overview'].apply(lambda x:x.split())
df_old=movies.copy()
```

```python
movies['tokens'] = movies['overview'].apply(lambda x:x.split()) + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
# make sentences from tokens to input to countVectorizer
movies['tokens'] = movies['tokens'].apply(lambda x: " ".join(x))
movies.head()
```

```python
df_old
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies['tokens']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
movies[movies['title'] == 'Spider-Man 3'].index[0]
```

```python
def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Spider-Man 3')
```
